**Підготовка**
Підключення бібліотек та завантаження даних

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/diamonds.csv')
```

**Завдання №1**
Побудувати стовпчикові діаграми, на яких відобразити а) кількість
діамантів кожного з класів якості; б) середню ціну діамантів
кожного класу якості; в) середню глибину діамантів різного класу
якості з різною якістю кольору.

```python
plt.figure()
plt.title('Count of all categories')
sns.countplot(data=data,x='cut')
plt.figure()
plt.title('Mean value of each category')
sns.barplot(x='cut',y='price',data=data)
plt.figure()
plt.title('Color\Deth')
sns.barplot(x='cut',y='depth',hue='color',data=data)
plt.show()
```

**Завдання №2**
Побудувати гістограму глибини діамантів у відсотках (depth),
загальну і для кожного класу якості.

```python
sns.histplot(x='depth',data=data)
plt.show()
```

```python
plt.figure()
sns.histplot(x='depth',hue='cut',data=data)
plt.figure()
sns.histplot(x='depth',hue='cut',multiple='stack',data=data)
plt.show()
```

**Завдання №3**
Побудувати діаграму розмаху параметру table (загальну і в
залежності від якості кольору), визначити чи присутні викиди.

```python
plt.subplot(1,2,1)
sns.boxplot(y='table',data=data)
plt.subplot(1,2,2)
sns.boxplot(x='color',y='table',data=data)
plt.show()
```

**Завдання №4**
За допомогою діаграм розсіювання зробити висновки щодо
залежності між а) довжиною і шириною; б) глибиною у % і
глибиною у мм.

```python
plt.figure()
sns.scatterplot(data = data, x = "x", y = "y")
plt.figure()
sns.scatterplot(data = data, x = "depth", y = "z")
plt.show()
```

а) Явно бачимо , що між довжиною та шириною існує пряма залежність.
б) Неможливо точно сказати чи існує залежність між глибиною (z - величиною розміру діаманту) та глибиною у %

```python
np.corrcoef(data['depth'],data['z'])
```

За коефіціентом кореляції бачимо , що дійсно явної залежності не існує (це є логічним , оскільки значення є відображенням схожих понять , хоча і не ідентичних)
