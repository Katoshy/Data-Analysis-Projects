**Підготовка**
Підключення бібліотек та завантаження даних

```python
import numpy as np
import pandas as pd
data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Version 1.xlsx')
cols = data.columns.str.split(',')
data = pd.DataFrame(data = [data.loc[row][0].split(',') for row in range(data.shape[0])],columns=cols[0])
print(data)
```

Заповнимо пусті значення

```python
print("Before dropping",data.shape[0])
data = data.drop_duplicates()
print("After dropping",data.shape[0])
data = data.replace('', np.nan)
data = data.replace(' ', np.nan)
data.fillna(0, inplace=True)
#data.dropna() thresh = 1,2,3...
#data.assign(wind_speed=lambda x:
#x.wind_speed.fillna(x.wind_speed.median()-1),meanpressure=lambda x:
#x.meanpressure.fillna(x.meanpressure.median()-1))
```

```python
print(data.dtypes)
data['date']=pd.to_datetime(data['date'])
for col in range(1,len(cols[0])):
  data[f'{cols[0][col]}'] = pd.to_numeric(data[f'{cols[0][col]}'])
print(data.dtypes)
```

```python
print(data)
data.rename(columns={'date':'Date'},inplace = True)
print(data)
Cols = [i.title() for i in cols[0]]
data.columns = Cols
print(data)
```

```python
print(data[data.Wind_Speed>=5].sort_values(by='Humidity',ascending=False))
```
